```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#OR
plt.show() #to display your plots when not using the Jupyter notebook
```

```python
import numpy as np
x = np.linspace(0, 10, 20)
y = x**2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.title('Our First Plot')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'red') 

plt.subplot(1,2,2)
plt.plot(x,y,'green')
```
